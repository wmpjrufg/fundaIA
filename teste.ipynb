```python
def data_comb(planilha, df, combinacoes):
    """
    Gera combinações 3 a 3 das colunas de uma planilha com header duplo,
    filtrando por combinações específicas fornecidas em um dicionário,
    e retorna uma lista de dicionários no formato desejado.

    Parâmetros:
    - planilha: Caminho para o arquivo Excel com header duplo.
    - combinacoes: Dicionário com combinações desejadas no formato 
      {'combinação 1': ['Fz', 'Mx', 'My'], ...}

    Retorno:
    - Uma lista de dicionários contendo os valores das combinações.
    """
    # df = pd.read_excel(planilha, header=[0, 1])

    lista_resultados = []
    for _, row in df.iterrows():
        resultado_linha = {}
        for nome_combinacao, colunas_desejadas in combinacoes.items():
            colunas_multiindex = [(nome_combinacao, coluna) for coluna in colunas_desejadas]
            if all(coluna in df.columns for coluna in colunas_multiindex):
                valores = row[colunas_multiindex].tolist()
                resultado_linha[nome_combinacao] = valores
            else:
                print(f"A combinação {colunas_desejadas} não está disponível para {nome_combinacao}.")
        
        lista_resultados.append(resultado_linha)

    return lista_resultados

combinacoes = {
    'combinação 1': ['Fz', 'Mx', 'My'],
    'combinação 2': ['Fz', 'Mx', 'My'],
    'combinação 3': ['Fz', 'Mx', 'My']
}

lista_comb = data_comb("planilha_padrao.xlsx", df_aux, combinacoes)
lista_comb
```

```python
dim_a = []
dim_b = []
df
for i, row in df.iterrows():
        # Recolhendo os dados do pilar
        dados_fundacao = { 
                                'combinações': lista_comb[i],
                                'sigma_adm (kPa)': row[('sigma_adm (kPa)', '')],
                         }
        # Otimização
        algorithm_setup = {   
                'number of iterations': 100,
                'number of population': 5,
                'number of dimensions': 2,
                'x pop lower limit': [0.60, 0.60], # 0.60
                'x pop upper limit': [2.25, 2.25],
                'none variable': dados_fundacao,
                'objective function': obj_ic_fundacoes,
                'algorithm parameters': {
                                        'selection': {'type': 'roulette'},
                                        'crossover': {'crossover rate (%)': 82, 'type':'linear'},
                                        'mutation': {'mutation rate (%)': 12, 'type': 'hill climbing', 'cov (%)': 15, 'pdf': 'gaussian'},
                                        }
                }
        n_rep = 5
        general_setup = {   
                'number of repetitions': n_rep,
                'type code': 'real code',
                'initial pop. seed': [None] * n_rep,
                'algorithm': 'genetic_algorithm_01',
                }
        df_all_reps, df_resume_all_reps, reports, status = metaheuristic_optimizer(algorithm_setup, general_setup)
        df_novo = df_resume_all_reps[status]
        dimensoes_sapata_a = list(df_novo['X_0_BEST'])[-1]
        dimensoes_sapata_b = list(df_novo['X_1_BEST'])[-1]
        dim_a.append(dimensoes_sapata_a)
        dim_b.append(dimensoes_sapata_b)

# Atribuição da dimensão otimizada
df['dimensão a (m)'] = dim_a
df['dimensão b (m)'] = dim_b
```

```python
df = df[[col for col in df.columns if not col[0].startswith("combinação")]]
df.columns = df.columns.get_level_values(0)
df
```
