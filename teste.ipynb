```python
import pandas as pd
from itertools import combinations

def combinar_dados(planilha, combinacoes):
    """
    Gera combinações 3 a 3 das colunas de uma planilha com header duplo,
    filtrando por combinações específicas fornecidas em um dicionário.

    Parâmetros:
    - planilha: Caminho para o arquivo Excel com header duplo.
    - combinacoes: Dicionário com combinações desejadas no formato 
      {'combinação 1': ['Fz', 'Mx', 'My'], ...}

    Retorno:
    - Um DataFrame contendo as combinações filtradas.
    """
    # Carregar a planilha Excel com cabeçalho duplo
    df = pd.read_excel(planilha, header=[0, 1])

    # Resultado final
    resultados = {}

    # Iterar sobre as combinações desejadas
    for nome_combinacao, colunas_desejadas in combinacoes.items():
        # Verificar se as colunas desejadas existem no MultiIndex
        colunas_multiindex = [(nome_combinacao, coluna) for coluna in colunas_desejadas]
        if all(coluna in df.columns for coluna in colunas_multiindex):
            # Filtrar as colunas desejadas
            resultados[nome_combinacao] = df[colunas_multiindex]
        else:
            print(f"A combinação {colunas_desejadas} não está disponível para {nome_combinacao}.")

    # Combinar todas as combinações em um único DataFrame
    df_combinado = pd.concat(resultados, axis=1)
    return df_combinado

# Exemplo de uso
combinacoes = {
    'combinação 1': ['Fz', 'My', 'My'],  # Substituir pelos nomes reais do nível inferior
    'combinação 3': ['Fz', 'Mx', 'My'],  # Pode repetir combinações
}

# Substituir "sua_planilha.xlsx" pelo caminho correto
resultado = combinar_dados("planilha_padrao.xlsx", combinacoes)
resultado
```
